```python
# Import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
# Load data

df = pd.read_csv('C:\Game\Proga\py.projects\machinelearning\Trojan.csv')
df.head()
```

```python
# Clean column names

df.columns = [i.strip() for i in df.columns]
```

```python
# Get info about the data

df.info()
```

```python
# Get summary statistics

df.describe()
```

```python
# Drop columns with only one unique value

df.drop(df.columns[df.nunique() == 1], axis=1, inplace=True)
```

```python
# Drop id column

df.drop('id', axis=1, inplace=True)
```

```python
# Find numerical and categorical features

numerical = df.select_dtypes(exclude=['object']).columns
categorical = df.select_dtypes(include=['object']).columns
numerical, categorical
```

```python
# Check unique values in the categorical features


{i: df[i].unique().size for i in categorical}
```

```python
# Encode target variable

from sklearn.preprocessing import LabelEncoder

df['Class'] = LabelEncoder().fit_transform(df['Class'])

X = df.drop('Class', axis=1)
numerical = X.select_dtypes(exclude=['object']).columns
categorical = X.select_dtypes(include=['object']).columns

y = df['Class'].astype(np.int64)
```

```python
# Plot correlation between numerical features

plt.figure(figsize=(20, 20))
sns.heatmap(X[numerical].corr(), annot=False, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Plot correlation between categorical features


plt.figure(figsize=(20, 20))
sns.heatmap(X[categorical].apply(lambda x: x.astype('category').cat.codes).corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Apply label encoding to the categorical features (because of the large number of categorical features)

from sklearn.preprocessing import LabelEncoder


X[categorical] = X[categorical].apply(LabelEncoder().fit_transform)
X[categorical].describe()
```

```python
# Plot correlation between all features

plt.figure(figsize=(20, 20))
sns.heatmap(X.corr(), annot=False, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Plot target variable 

plt.figure(figsize=(10, 10))
y.plot.hist()
plt.show()
```

```python
# Split data into train and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42, test_size=0.2) 

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Check for missing values

X_train.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

X_train = SimpleImputer(strategy='most_frequent').fit_transform(X_train)
X_train = MinMaxScaler().fit_transform(X_train)

X_test = SimpleImputer(strategy='most_frequent').fit_transform(X_test)
X_test = MinMaxScaler().fit_transform(X_test)
```

```python
# Apply kNN

from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier(**{'metric': 'euclidean', 'n_neighbors': 80, 'weights': 'distance'})
knn.fit(X_train, y_train)
```

```python
# Apply kNN on test set

y_pred = knn.predict(X_test)
```

```python
# Check metrics

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(f1_score(y_test, y_pred))
```

```python
# Plot confusion matrix

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['benign', 'malicious'],
            yticklabels=['benign', 'malicious'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
